```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('/content/pizza_transaction.csv')
```

```python
df.head(3)
df.info()
```

```python
# cheak missing values
df.isnull().sum()
```

```python
# Change 'order_time', 'order_date' to be Data_type : Datetime
df['order_time'] = pd.to_datetime(df['order_time'])
df['order_date'] = pd.to_datetime(df['order_date'])
df.info()
```

```python
# create new column : 'hour'
df['hour'] = df['order_time'].dt.hour
df['hour'].head()
# df.info()

# create new column : 'date_name'
df['date_name'] = df['order_date'].dt.day_name()
df['date_name'].head()
df.info()
```

```python

# create new column : 'month_name'
df['month_name'] =pd.DatetimeIndex (df['order_date']).month
df.loc[(df['month_name'] ==1), 'month_name'] = 'January'
df.loc[(df['month_name'] ==2), 'month_name'] = 'February'
df.loc[(df['month_name'] ==3), 'month_name'] = 'March'
df.loc[(df['month_name'] ==4), 'month_name'] = 'April'
df.loc[(df['month_name'] ==5), 'month_name'] = 'May'
df.loc[(df['month_name'] ==6), 'month_name'] = 'June'
df.loc[(df['month_name'] ==7), 'month_name'] = 'July'
df.loc[(df['month_name'] ==8), 'month_name'] = 'August'
df.loc[(df['month_name'] ==9), 'month_name'] = 'September'
df.loc[(df['month_name'] ==10), 'month_name'] = 'October'
df.loc[(df['month_name'] ==11), 'month_name'] = 'November'
df.loc[(df['month_name'] ==12), 'month_name'] = 'December'
df['month_name'].value_counts()
# df.head()
df.info()
```

```python
# create variable for card chart

# 1 : Total Revenue
total_sales = (df['total_price']).sum()
print("Total Revenue:", total_sales)

# 2 : Total Orders
total_orders = len(df.groupby('order_id').count())
print("Total Orders:", total_orders)

# 3 : Average Sales per month
num_months = df['month_name'].nunique()
print("Number of Months:", num_months)
avg_sales_per_month = total_sales / num_months
print("Average Sales per Month:", avg_sales_per_month)

# 4 : Total Pizza Sales
total_products = (df['quantity']).sum()
print("Total Pizza Sales:", total_products)

# 5 : Pizza Category
total_category = df.category.nunique()
print("Total Category:", total_category)

# 6 : Pizza Menu
# total_pizza = len(df.groupby('pizza_id').count())
total_pizza = df.name.nunique()
print("Pizza Menu:", total_pizza)


print("-"*20)
total_products = (df['quantity']).sum()
print("Total Pizza Sales:", total_products)

avg_orders_per_month = total_orders / num_months
print("Average Orders per Month:", avg_orders_per_month)

unique_dates = df['order_date'].unique()
total_days = len(unique_dates)
print("Total Days:", total_days)

# total_days = df['order_date'].nunique()
# print("Total Days:", total_days)

avg_orders_per_day = total_orders / total_days
print("Average Orders per Day:", avg_orders_per_day)

```

```python
# create data_table to create value in card chart
data_table = {
    'Total Revenue': str(f'{int(total_sales):,}')+' $',
    'Total Orders' : f'{total_orders:,}',
    'Pizza Category': f'{total_category:,}',
    'Average Sales per Month': str(f'{int(avg_sales_per_month):,}')+' $',
    'Yearly Pizza Volume' : f'{int(total_products):,}',
    'Pizza Menu' : total_pizza
}

data_table
```

```python
# create new table : total_sales_months
total_sales_months = df.groupby('month_name').agg(
    sales_per_month =('total_price', 'sum'),
    order_per_month =('order_id', 'nunique')
    )

total_sales_months = total_sales_months.reset_index()

# Before sort by month_order
total_sales_months

month_order = ['january', 'february', 'march', 'april', 'may', 'june',
               'july', 'august', 'september', 'october', 'november', 'december']

# loop for change to be capitalize
month_order = [month.capitalize() for month in month_order]

# create dataframe order by month_order
total_sales_months['month_name'] = pd.Categorical(total_sales_months['month_name'], categories=month_order, ordered=True)

# After sorting by month_order
total_sales_months = total_sales_months.sort_values('month_name')

total_sales_months
```

```python
# check Years in dataset
df['order_date'].dt.year.unique()
```

```python
# create new column : average_sales_per_month

total_sales_months['average_sales_per_month'] = avg_sales_per_month
total_sales_months
```

```python
# create chart : Monthly Sales Chart
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib import gridspec

# Data Frame
total_sales_months

# Set the proportions of the subplot.
gs = gridspec.GridSpec(1, 2, width_ratios=[4, 1])  # left:right = 4:1
fig, axes = plt.subplots(ncols=2, nrows=1, figsize=(15, 6), gridspec_kw={'width_ratios': [4, 1]})

# Left Chart (Bar chart)
ax1 = axes[0]

# color by Max and Min sales month
max_sales_month = total_sales_months.loc[total_sales_months['sales_per_month'].idxmax()]
min_sales_month = total_sales_months.loc[total_sales_months['sales_per_month'].idxmin()]

for index, row in total_sales_months.iterrows():
    color = '#4db6ac'
    if row['month_name'] == max_sales_month['month_name']:
        color = '#6b9e2e'
    elif row['month_name'] == min_sales_month['month_name']:
        color = '#d3e65b'
    ax1.bar(row['month_name'], row['sales_per_month'], color=color)

# set label at bar chart
for x, y in zip(total_sales_months['month_name'], total_sales_months['sales_per_month']):
    ax1.text(x, y, str(f'{int(y):,}'), ha='center', va='bottom', fontsize=11)

# set xlabel, ylabel, title
ax1.set_xlabel('Chart no.1 : Sales by Month in 2015', fontsize=18,fontweight='bold',labelpad=10)
ax1.set_ylabel('Sales', color='black', fontsize=12)
ax1.set_title('Monthly Sales ', fontsize= 20, fontweight='bold', loc='center',pad=10)

ax1.tick_params(axis='y', labelsize=12)
ax1.tick_params(axis='x', labelsize=12, rotation=45)

# Right Chart (Card chart)
ax_card = axes[1]
ax_card.set_facecolor('#f0f0f0')
ax_card.spines['top'].set_visible(True)
ax_card.spines['right'].set_visible(True)
ax_card.spines['bottom'].set_visible(True)
ax_card.spines['left'].set_visible(True)
ax_card.set_xticks([])
ax_card.set_yticks([])

total_sales = total_sales_months['sales_per_month'].sum()

# add text in Card chart
ax_card.text(0.5, 0.6, 'Total Sales:', ha='center', fontsize=20, fontweight='bold', color='#333333')
ax_card.text(0.5, 0.4, str(f'{int(total_sales):,}')+" $", ha='center', fontsize=24, color='#333333')


plt.tight_layout()


plt.show()
```
