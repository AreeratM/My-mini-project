```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
order_details = pd.read_csv('/content/order_details.csv')
order_details.head(3)
order_details.isnull().sum()
```

```python
orders = pd.read_csv('/content/orders.csv')
orders.head(3)
orders.isnull().sum()
```

```python
pizza_types = pd.read_csv('/content/pizza_types.csv')
pizza_types.head(3)
pizza_types.isnull().sum()
```

```python
pizzas = pd.read_csv('/content/pizzas.csv')
pizzas.head(3)
pizzas.isnull().sum()
```

```python
pizza_1 = order_details.merge(orders, left_on='order_id',right_on='order_id')
pizza_1.head(3)
pizza_2 = pizza_1.merge(pizzas, left_on='pizza_id',right_on='pizza_id')
pizza_2.head(3)
pizza_4 = pizza_2.merge(pizza_types, left_on='pizza_type_id',right_on='pizza_type_id')
pizza_4.head(3)
```

```python
# pizza_4 = pizza_4.drop(columns=['pizza_type_id'])
pizza_4.info()
```

```python
pizza_4.rename(columns= {'date' : 'order_date',
                         'time' : 'order_time',
                         'price' : 'unit_price',
                         'size' : 'pizza_size',}, inplace=True)
# pizza_4.rename(columns={'date' : 'order_date'}, inplace=True)
```

```python
# pizza_4.head(3)
pizza_4.info()
```

```python
pizza_4 = pizza_4.drop(columns=['pizza_type_id','order_details_id'])
```

```python
pizza_4['total_price'] = pizza_4['quantity'] * pizza_4['unit_price']
```

```python
pizza_4.head(3)
```

```python
pizza_4 = pizza_4.reindex(columns=['order_id', 'pizza_id','order_date', 'order_time', 'name', 'pizza_size',
                         'unit_price', 'quantity','total_price','category','ingredients'])
```
